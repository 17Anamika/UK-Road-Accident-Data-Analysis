### 4.Decoding the coded columns

```python
import pandas as pd

df = pd.read_csv("collision-2024.csv")
guide_df = pd.read_csv("guide_file.csv")

# Remove accidental duplicates in guide and strip whitespace
guide_df = guide_df.drop_duplicates(subset=["field name","code"])
guide_df['code'] = guide_df['code'].astype(str).str.strip()
guide_df['label'] = guide_df['label'].astype(str).str.strip()


coded_columns = guide_df['field name'].unique()
coded_columns = [col for col in coded_columns if col in df.columns]

print(coded_columns)
```

```python
for col in coded_columns:
    mapping = guide_df[guide_df['field name'] == col][["code","label"]]
    map_dict = pd.Series(mapping['label'].values, index=mapping['code']).to_dict()
    # convert df values to str for mapping compatibility
    df[col] = df[col].astype(str).str.strip().map(map_dict).fillna(df[col])
    print("after decoding shape:", df.shape)



df.to_csv("collisions_decoded7.csv", index=False)
```

### 1.6 Feature Engineering: Risk Rate Features for Multiple Columns

Risk rate features calculate the historical probability of fatal/serious collisions for each category. These are powerful features for analysis, SQL queries, and Power BI dashboards.


```python
# CREATING RISK RATE FEATURES FOR MULTIPLE CATEGORICAL COLUMNS
# These features tell us: "How risky is this category historically?"

print("=" * 70)
print("CREATING RISK RATE FEATURES")
print("=" * 70)

# List of columns for which we want to calculate risk rates
columns_for_risk_rates = [
    'road_type',
    'junction_detail', 
    'weather_conditions',
    'road_surface_conditions',
    'urban_or_rural_area',
    'light_conditions',
    'speed_limit',
    'first_road_class',
    'season',
    'geo_cell',
    'weekend',
    'time_of_day'
]

print(f"\nColumns for risk rate calculation: {columns_for_risk_rates}")
print(f"\nNote: We'll calculate the fatal_or_serious_rate for each category")
```

```python
# FUNCTION TO CALCULATE RISK RATE FOR ANY COLUMN
def calculate_risk_rate(df, column_name, severity_flag='is_fatal_or_serious'):
    """
    Calculate the risk rate (fatal/serious rate) for each category in a column.
    
    Parameters:
    - df: DataFrame
    - column_name: Name of the categorical column
    - severity_flag: Binary flag column to use (default: 'is_fatal_or_serious')
    
    Returns:
    - DataFrame with category and its risk rate
    """
    risk_stats = df.groupby(column_name).agg({
        severity_flag: ['count', 'sum', 'mean']
    }).reset_index()
    
    # Flatten column names
    risk_stats.columns = [column_name, 'total_count', 'fatal_serious_count', 'fatal_serious_rate']
    
    # Add percentage version (rate * 100) for easier interpretation
    risk_stats['fatal_serious_rate_pct'] = risk_stats['fatal_serious_rate'] * 100
    
    return risk_stats

# Test the function with one column first
print("Example: Risk rate calculation for 'road_type':")
road_type_risk = calculate_risk_rate(df, 'road_type')
print(road_type_risk.sort_values('fatal_serious_rate', ascending=False).head(10))
```

```python
# CALCULATE RISK RATES FOR ALL COLUMNS AND MERGE BACK TO DATAFRAME

# Create a dictionary to store all risk rate dataframes
risk_rate_dict = {}

# Calculate risk rates for each column
for col in columns_for_risk_rates:
    if col in df.columns:
        risk_df = calculate_risk_rate(df, col)
        # Store only the category and rate columns for merging
        risk_rate_dict[col] = risk_df[[col, 'fatal_serious_rate']].copy()
        risk_rate_dict[col].columns = [col, f'{col}_fatal_serious_rate']
        
        print(f"✅ Calculated risk rate for: {col}")
        print(f"   Categories: {len(risk_df)}")
    else:
        print(f"⚠️  Column '{col}' not found in dataframe")

print(f"\nTotal risk rate features created: {len(risk_rate_dict)}")
```

### 1. Chi-Square Tests: Testing Associations Between Categorical Variables

**Purpose**: Test if there's a statistically significant association between two categorical variables.

**Example Questions**:
- Is collision severity significantly associated with weather conditions?
- Is severity significantly different across road types?
- Does weekend vs weekday affect severity distribution?

**Interpretation**:
- **p < 0.05**: Significant association (not due to chance)
- **p >= 0.05**: No significant association (could be random)


```python
# CHI-SQUARE TEST: Collision Severity vs Various Categorical Factors

def chi_square_test(df, var1, var2, alpha=0.05):
    """
    Perform chi-square test of independence between two categorical variables.
    
    Parameters:
    - df: DataFrame
    - var1: First categorical variable (e.g., 'collision_severity')
    - var2: Second categorical variable (e.g., 'weather_conditions')
    - alpha: Significance level (default 0.05)
    
    Returns:
    - Dictionary with test results
    """
    # Create contingency table
    contingency_table = pd.crosstab(df[var1], df[var2])
    
    # Perform chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    
    # Determine significance
    is_significant = p_value < alpha
    
    results = {
        'chi2_statistic': chi2,
        'p_value': p_value,
        'degrees_of_freedom': dof,
        'is_significant': is_significant,
        'contingency_table': contingency_table,
        'interpretation': 'Significant association' if is_significant else 'No significant association'
    }
    
    return results

print("=" * 70)
print("CHI-SQUARE TESTS: Severity vs Categorical Factors")
print("=" * 70)

# List of categorical factors to test against severity
factors_to_test = [
    'weather_conditions',
    'road_surface_conditions',
    'light_conditions',
    'urban_or_rural_area',
    'weekend',
    'speed_band',
    'season',
    'geo_cell',
    'junction_detail',
    'road_type',
    'pedestrian_crossing',
    'police_force'  # Note: speed_limit might need to be categorical
]

# Perform chi-square tests
chi_square_results = {}

for factor in factors_to_test:
    if factor in df.columns:
        try:
            result = chi_square_test(df, 'collision_severity', factor)
            chi_square_results[factor] = result
            
            print(f"\n{factor.upper().replace('_', ' ')} vs Collision Severity:")
            print(f"  Chi-square statistic: {result['chi2_statistic']:.4f}")
            print(f"  p-value: {result['p_value']:.6f}")
            print(f"  Degrees of freedom: {result['degrees_of_freedom']}")
            print(f"  Result: {result['interpretation']} (α = 0.05)")
            
        except Exception as e:
            print(f"\n⚠️  Error testing {factor}: {str(e)}")
    else:
        print(f"\n⚠️  Column '{factor}' not found in dataframe")
```

```python
# DISPLAY DETAILED RESULTS OF CHI-SQUARE TESTS

print("=" * 70)
print("DETAILED CHI-SQUARE TEST RESULTS SUMMARY")
print("=" * 70)


# Create a summary dataframe
summary_data = []
for factor, result in chi_square_results.items():
    summary_data.append({
        'Factor': factor.replace('_', ' ').title(),
        'Chi-square': result['chi2_statistic'],
        'p-value': result['p_value'],
        'Significant': 'Yes' if result['is_significant'] else 'No',
        'Interpretation': result['interpretation']
    })

chi_summary_df = pd.DataFrame(summary_data)
chi_summary_df = chi_summary_df.sort_values('p-value')

print("\nTest Results (sorted by p-value):")
print(chi_summary_df.to_string(index=False))

print("\n" + "=" * 70)
print("INTERPRETATION GUIDE:")
print("=" * 70)
print("  p < 0.001: Very strong evidence of association (***)")
print("  p < 0.01:  Strong evidence of association (**)")
print("  p < 0.05:  Evidence of association (*)")
print("  p >= 0.05: No significant association")
print("=" * 70)
```

### 2. ANOVA / Kruskal-Wallis Tests: Comparing Numeric Variables Across Groups

**Purpose**: Test if the mean/median of a numeric variable differs significantly across groups.

**Example Questions**:
- Do average casualties differ significantly across speed bands?
- Are there differences in vehicle counts across weather conditions?
- Does the number of casualties differ between urban and rural areas?

**Tests**:
- **ANOVA (f_oneway)**: Parametric test (assumes normal distribution)
- **Kruskal-Wallis**: Non-parametric alternative (no distribution assumptions)

**Interpretation**:
- **p < 0.05**: At least one group is significantly different
- **p >= 0.05**: No significant differences between groups


```python
# ANOVA / KRUSKAL-WALLIS TESTS: Comparing Numeric Variables Across Groups

def compare_groups_numeric(df, numeric_var, group_var, use_parametric=True):
    """
    Compare numeric variable across groups using ANOVA or Kruskal-Wallis.
    
    Parameters:
    - df: DataFrame
    - numeric_var: Name of numeric variable (e.g., 'number_of_casualties')
    - group_var: Name of categorical grouping variable (e.g., 'speed_band')
    - use_parametric: If True, use ANOVA; if False, use Kruskal-Wallis
    
    Returns:
    - Dictionary with test results
    """
    # Get groups
    groups = [group_df[numeric_var].dropna().values 
              for name, group_df in df.groupby(group_var)]
    
    # Remove empty groups
    groups = [g for g in groups if len(g) > 0]
    
    if len(groups) < 2:
        return None
    
    # Perform test
    if use_parametric:
        statistic, p_value = f_oneway(*groups)
        test_name = 'ANOVA (F-test)'
    else:
        statistic, p_value = kruskal(*groups)
        test_name = 'Kruskal-Wallis'
    
    is_significant = p_value < 0.05
    
    results = {
        'test_name': test_name,
        'statistic': statistic,
        'p_value': p_value,
        'is_significant': is_significant,
        'interpretation': 'Significant difference between groups' if is_significant else 'No significant difference'
    }
    
    return results

print("=" * 70)
print("ANOVA / KRUSKAL-WALLIS TESTS")
print("=" * 70)

# Tests to perform: (numeric_variable, categorical_grouping_variable)
tests_to_perform = [
    ('number_of_casualties', 'speed_band'),
    ('number_of_casualties', 'weather_conditions'),
    ('number_of_casualties', 'urban_or_rural_area'),
    ('number_of_casualties', 'road_surface_conditions'),
    ('number_of_vehicles', 'road_surface_conditions'),
    ('number_of_casualties', 'light_conditions'),
    ('number_of_vehicles', 'light_conditions'),
    ('number_of_vehicles', 'urban_or_rural_area'),
    ('number_of_vehicles', 'weather_conditions'),
    ('number_of_vehicles', 'speed_band'),
    ('number_of_casualties', 'weekend'),
    ('number_of_vehicles', 'weekend'),
    ('casualties_per_vehicle', 'season'),
    ('casualties_per_vehicle', 'geo_cell'),
    ('casualties_per_vehicle', 'junction_detail'),
    ('casualties_per_vehicle', 'police_force'),
    ('casualties_per_vehicle', 'time_of_day')
]

anova_results = {}

for numeric_var, group_var in tests_to_perform:
    if numeric_var in df.columns and group_var in df.columns:
        # Check if group_var has multiple categories
        if df[group_var].nunique() >= 2:
            try:
                # Try both parametric and non-parametric
                result_anova = compare_groups_numeric(df, numeric_var, group_var, use_parametric=True)
                result_kruskal = compare_groups_numeric(df, numeric_var, group_var, use_parametric=False)
                
                if result_anova and result_kruskal:
                    key = f"{numeric_var} vs {group_var}"
                    anova_results[key] = {
                        'anova': result_anova,
                        'kruskal': result_kruskal
                    }
                    
                    print(f"\n{numeric_var.upper().replace('_', ' ')} vs {group_var.upper().replace('_', ' ')}:")
                    print(f"  ANOVA F-statistic: {result_anova['statistic']:.4f}, p-value: {result_anova['p_value']:.6f}")
                    print(f"  Kruskal-Wallis H-statistic: {result_kruskal['statistic']:.4f}, p-value: {result_kruskal['p_value']:.6f}")
                    print(f"  Result: {result_kruskal['interpretation']} (α = 0.05)")
                    
            except Exception as e:
                print(f"\n⚠️  Error testing {numeric_var} vs {group_var}: {str(e)}")
        else:
            print(f"\n⚠️  {group_var} has less than 2 categories")
    else:
        missing = [v for v in [numeric_var, group_var] if v not in df.columns]
        print(f"\n⚠️  Column(s) not found: {missing}")
```

```python
# SUMMARY OF ANOVA/KRUSKAL-WALLIS RESULTS

print("=" * 70)
print("SUMMARY: ANOVA/KRUSKAL-WALLIS TEST RESULTS")
print("=" * 70)

summary_data = []
for test_name, results in anova_results.items():
    # Use Kruskal-Wallis result (more robust)
    kruskal_result = results['kruskal']
    anova_result = results['anova']
    
    summary_data.append({
        'Comparison': test_name.replace(' vs ', ' vs\n'),
        'Kruskal-Wallis H': kruskal_result['statistic'],
        'p-value': kruskal_result['p_value'],
        'Significant': 'Yes' if kruskal_result['is_significant'] else 'No'
    })

anova_summary_df = pd.DataFrame(summary_data)
anova_summary_df = anova_summary_df.sort_values('p-value')

print("\nTest Results (sorted by p-value):")
print(anova_summary_df.to_string(index=False))
```

```python
# CORRELATION TESTS: Numeric Variables

def correlation_test(df, var1, var2):
    """
    Perform Pearson and Spearman correlation tests.
    
    Parameters:
    - df: DataFrame
    - var1: First numeric variable
    - var2: Second numeric variable
    
    Returns:
    - Dictionary with correlation results
    """
    # Remove missing values
    data = df[[var1, var2]].dropna()
    
    if len(data) < 3:
        return None
    
    # Pearson correlation
    pearson_r, pearson_p = pearsonr(data[var1], data[var2])
    
    # Spearman correlation
    spearman_rho, spearman_p = spearmanr(data[var1], data[var2])
    
    # Interpret strength
    def interpret_correlation(r):
        abs_r = abs(r)
        if abs_r > 0.7:
            return 'Strong'
        elif abs_r > 0.3:
            return 'Moderate'
        else:
            return 'Weak'
    
    results = {
        'pearson_r': pearson_r,
        'pearson_p': pearson_p,
        'pearson_interpretation': interpret_correlation(pearson_r),
        'pearson_significant': pearson_p < 0.05,
        'spearman_rho': spearman_rho,
        'spearman_p': spearman_p,
        'spearman_interpretation': interpret_correlation(spearman_rho),
        'spearman_significant': spearman_p < 0.05,
        'n_observations': len(data)
    }
    
    return results

print("=" * 70)
print("CORRELATION TESTS: Numeric Variables")
print("=" * 70)

# Pairs of numeric variables to test
correlation_pairs = [
    ('number_of_vehicles', 'number_of_casualties'),
    ('longitude', 'latitude'),  # Geographic correlation (usually weak)
]

# Also test correlations with binary flags
correlation_pairs_with_flags = [
    ('number_of_vehicles', 'is_fatal_or_serious'),
    ('number_of_casualties', 'is_fatal_or_serious'),
    ('number_of_vehicles', 'is_slight'),
    ('number_of_casualties', 'is_slight')
]

correlation_results = {}

# Test numeric-numeric correlations
print("\n1. NUMERIC VARIABLE CORRELATIONS:")
for var1, var2 in correlation_pairs:
    if var1 in df.columns and var2 in df.columns:
        result = correlation_test(df, var1, var2)
        if result:
            correlation_results[f"{var1} vs {var2}"] = result
            
            print(f"\n{var1.upper().replace('_', ' ')} vs {var2.upper().replace('_', ' ')}:")
            print(f"  Pearson r: {result['pearson_r']:.4f}, p-value: {result['pearson_p']:.6f}")
            print(f"    → {result['pearson_interpretation']} correlation, {'Significant' if result['pearson_significant'] else 'Not significant'}")
            print(f"  Spearman ρ: {result['spearman_rho']:.4f}, p-value: {result['spearman_p']:.6f}")
            print(f"    → {result['spearman_interpretation']} correlation, {'Significant' if result['spearman_significant'] else 'Not significant'}")
            print(f"  Observations: {result['n_observations']}")

# Test correlations with binary flags (point-biserial correlation)
print("\n2. CORRELATIONS WITH BINARY FLAGS:")
for var1, var2 in correlation_pairs_with_flags:
    if var1 in df.columns and var2 in df.columns:
        result = correlation_test(df, var1, var2)
        if result:
            correlation_results[f"{var1} vs {var2}"] = result
            
            print(f"\n{var1.upper().replace('_', ' ')} vs {var2.upper().replace('_', ' ')}:")
            print(f"  Pearson r: {result['pearson_r']:.4f}, p-value: {result['pearson_p']:.6f}")
            print(f"    → {result['pearson_interpretation']} correlation, {'Significant' if result['pearson_significant'] else 'Not significant'}")
            print(f"  Spearman ρ: {result['spearman_rho']:.4f}, p-value: {result['spearman_p']:.6f}")
            print(f"    → {result['spearman_interpretation']} correlation, {'Significant' if result['spearman_significant'] else 'Not significant'}")
```

```python

# This script pre-processes your CSV to handle common datatype errors

import pandas as pd
import numpy as np

# ===== STEP 1: Read your CSV file =====
csv_file = 'collisions_final_analysis.csv'  # Change to your CSV file
df_for_sql = pd.read_csv(csv_file)

print(f"📖 Loaded CSV: {len(df_for_sql):,} rows, {len(df_for_sql.columns)} columns")
print(f"\n🔍 Checking data types and issues...")

# ===== STEP 2: Clean and Convert DataTypes =====
# Fix common issues that cause SQL Server import errors

# Replace NaN/None with empty strings for string columns, or keep NULL for numeric
df_cleaned = df_for_sql.copy()

# Handle common problematic characters
for col in df_cleaned.columns:
    # Convert object/string columns
    if df_cleaned[col].dtype == 'object':
        # Replace newlines, tabs, and other control characters that break CSV import
        df_cleaned[col] = df_cleaned[col].astype(str).replace(['\n', '\r', '\t'], [' ', ' ', ' '], regex=True)
        # Replace empty strings with None for NULL handling
        df_cleaned[col] = df_cleaned[col].replace(['nan', 'None', 'NaN', ''], None)
        
    # Convert numeric columns - handle errors gracefully
    elif df_cleaned[col].dtype in ['int32','int64', 'Int64']:
        # Keep integers as is, but handle NaN
        df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce').astype('Int64')  # Nullable integer
        
    elif df_cleaned[col].dtype in ['float64']:
        # Keep floats, but round very large numbers
        df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')

# ===== STEP 3: Create SQL-Safe CSV (for SSMS Import Wizard) =====
output_file = csv_file.replace('.csv', '_sql_ready.csv')
df_cleaned.to_csv(output_file, index=False, encoding='utf-8-sig', na_rep='NULL', quotechar='"', escapechar='\\')
print(f"\n✅ Created SQL-ready CSV: {output_file}")
print(f"   - Replaced problematic characters")
print(f"   - Handled NULL values")
print(f"   - UTF-8 encoding with BOM (works better with SSMS)")

```
